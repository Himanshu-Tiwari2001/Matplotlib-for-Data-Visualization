```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[45,34,56,23,45]
plt.subplot(2,2,1)
plt.title("Age")
plt.plot(x,y)

x=[5,6,7,8,9]
y=[67,50,66,56,82]
plt.subplot(2,2,2)
plt.title("Weight")
plt.bar(x,y)

x=[1,2,3,4,5]
y=[4500,3400,5600,2300,4500]
plt.subplot(2,2,3)
plt.title("Salary")
plt.scatter(x,y)

x=[5,6,7,8,9]
y=[670,500,660,560,820]
plt.subplot(2,2,4)
plt.title("Bonus")
plt.step(x,y)
plt.suptitle("Employee Data") #super title
plt.savefig("Employee data.png")
plt.show()
```
