```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("dataSets/expense3.xlsx")
df=pd.DataFrame(data)
print(df)
```

```python
grouped_by=df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
```

```python
grouped_by=df.groupby("Category")["Amount"].sum()
print(grouped_by)
# plt.bar(grouped_by.Category,grouped_by.Amount)
```

```python
 # Grouping the data
grouped_by = df.groupby("Category")["Amount"].sum()

# Plotting directly with Pandas
colors=["red","green","yellow","blue","orange" ,"pink"]
grouped_by.plot(kind="bar", figsize=(10, 6), color=colors)

plt.xlabel("Category",fontsize=15)

plt.ylabel("Total Amount",fontsize=15)
plt.title("Total Expenses by Category",fontsize=20)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
grouped_by.plot(kind="pie", figsize=(8, 10), color=colors)
plt.show()
grouped_by.plot(kind="line", figsize=(8, 10), color=colors,marker="o")
plt.show()



```
