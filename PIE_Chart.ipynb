```python
import  matplotlib.pyplot as plt
brand=["One Plus","Vivo","Google Pixel","Mi","Moto","Samsung"]
x=[12,35,30,3,10,10]
c=["Red","Silver","Blue","Magenta","Orange","Violet"]

ex=[0,0,0.1,0,0,0.1]  #it is used to highlight the particular data
plt.pie(x,labels=brand,colors=c,explode=ex,shadow=True,)
plt.show
```

```python
plt.pie(x,labels=brand,colors=c,explode=ex,shadow=False,autopct="%.2f")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel("dataSets/expense3.xlsx")
df=pd.DataFrame(data)
grouped_by=df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
plt.pie(grouped_by.values,labels=grouped_by.index,autopct="%.2f")
plt.show()
```
