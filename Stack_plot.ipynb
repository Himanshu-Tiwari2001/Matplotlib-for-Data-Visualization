```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6]

NOP1=[2,10,30,30,50,30]
NOP2=[2,15,35,39,55,35]
NOP3=[2,18,38,30,50,30]
plt.stackplot(days,NOP1,NOP2,NOP3,labels=["Week1","Week2","Week3"])
plt.legend()
plt.savefig("StackPlot.png")
plt.show()
#to save the plot as png as pdf etc


```
